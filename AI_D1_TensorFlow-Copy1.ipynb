```python
import tensorflow as tf
```

```python
# 1.Load DataSets
mnist = tf.keras.datasets.mnist     #(mnist = blury image)
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
# 2.Normalize Pixel values to range[0,1]
x_train = x_train / 255.0     #( 255 = array)
x_test = x_test / 255.0
```

```python
# 3.Flatten 28*28 image into 784-length vextions
x_train = x_train.reshape(-1,28*28)     #(28*28 = pixle)
x_test = x_test.reshape(-1,28*28)
```

```python
# 4.Build a Simple Neural Network
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128,activation = 'relu'),     #(Dense(single node) = layer)  #(relu = 0,1 form of image)  #(128 = input)
    tf.keras.layers.Dense(10,activation = 'softmax')     #(10 = input)    #(softmax = 0,1 form of image)
])
```

```python
# 5.Compile the model
model.compile(optimizer = 'adam',    #(adam = Gradient Descent)
              loss = 'sparse_categorical_crossentropy',    #(loss = MSE(mean score error))
              metrics = ['accuracy'])
```

```python
# 6.Train the model
model.fit(x_train,y_train,epochs = 5)    #(epochs = no. of times train the model)
```

```python
# 7.Evaluate the model
test_loss,test_acc = model.evaluate(x_test,y_test)
print(f"\n test accuracy : {test_acc:.4f}")
```
